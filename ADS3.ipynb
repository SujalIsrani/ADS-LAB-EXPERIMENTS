```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from pandas.plotting import scatter_matrix
```

```python
data = {
    "Student_ID": range(1, 16),
    "Marks": [56, 78, 45, 89, 67, 72, 88, 90, 54, 61, 70, 85, 76, 69, 73],
    "Study_Hours": [2.5, 4.0, 1.8, 5.2, 3.1, 3.5, 5.0, 5.5, 2.2, 2.8, 3.4, 4.8, 4.0, 3.0, 3.6],
    "Attendance": [75, 85, 65, 92, 80, 82, 90, 95, 70, 78, 84, 88, 86, 80, 83]
}

df = pd.DataFrame(data)
df.head()
```

```python
plt.hist(df['Marks'], bins=5)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(df['Student_ID'], df['Marks'])
plt.title("Bar Chart of Student Marks")
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.show()
```

```python
plt.boxplot(df['Marks'])
plt.title("Quartile Plot of Marks")
plt.ylabel("Marks")
plt.show()
```

```python
sns.kdeplot(df['Marks'], fill=True)
plt.title("Distribution Chart of Marks")
plt.show()
```

```python
plt.scatter(df['Study_Hours'], df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Scatter Plot: Study Hours vs Marks")
plt.show()
```

```python
scatter_matrix(df[['Marks','Study_Hours','Attendance']], figsize=(8,8))
plt.show()
```

```python
px.scatter(
    df,
    x="Study_Hours",
    y="Marks",
    size="Attendance",
    color="Attendance",
    title="Bubble Chart: Marks vs Study Hours"
)
```

```python
sns.kdeplot(df['Marks'], label="Marks")
sns.kdeplot(df['Attendance'], label="Attendance")
plt.legend()
plt.title("Density Chart")
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Heat Map of Correlation Matrix")
plt.show()
```
